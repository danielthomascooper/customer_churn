```python
test_df = pd.read_csv('/kaggle/input/competitions/playground-series-s6e3/test.csv')
test_df.head()
```

```python
# ============================================================================
# PRE-PROCESSING PIPELINE
# ============================================================================

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import warnings
warnings.filterwarnings('ignore')

# Step 1: Create a preprocessing function that handles categorical encoding consistently
def preprocess_data(df, is_training=True, encoder=None, scaler=None, target_col='Churn'):
    """
    Preprocess customer churn data with consistent encoding for train/test sets.
    
    Parameters:
    -----------
    df : pd.DataFrame
        Input dataframe
    is_training : bool
        Whether this is training data (True) or test data (False)
    encoder : dict
        Dictionary of encoders from training set (for test data)
    scaler : StandardScaler
        Fitted scaler from training set (for test data)
    target_col : str
        Name of target column (only in training data)
    
    Returns:
    --------
    X : pd.DataFrame
        Preprocessed features
    y : pd.Series (only if is_training=True)
        Target variable
    encoder : dict (only if is_training=True)
        Dictionary of encoders for reuse
    scaler : StandardScaler (only if is_training=True)
        Fitted scaler for reuse
    """
    
    df_processed = df.copy()
    
    # Identify feature columns (exclude id and target)
    feature_cols = [col for col in df_processed.columns if col not in ['id', target_col]]
    
    # Separate numeric and categorical features
    numeric_cols = df_processed[feature_cols].select_dtypes(include=[np.number]).columns.tolist()
    categorical_cols = df_processed[feature_cols].select_dtypes(include=['object']).columns.tolist()
    
    # Handle categorical variables with one-hot encoding
    if is_training:
        # For training data, fit the encoder
        df_dummies = pd.get_dummies(df_processed[categorical_cols], drop_first=True)
        encoder = {col: df_processed[col].unique() for col in categorical_cols}
    else:
        # For test data, ensure all categories match training data
        df_dummies = pd.get_dummies(df_processed[categorical_cols], drop_first=True)
    
    # Combine numeric and encoded categorical features
    X = pd.concat([df_processed[numeric_cols].reset_index(drop=True), 
                   df_dummies.reset_index(drop=True)], axis=1)
    
    # Handle missing values (fill with median for numeric, mode for categorical)
    if is_training:
        numeric_means = X[numeric_cols].mean()
        X[numeric_cols] = X[numeric_cols].fillna(numeric_means)
        # Fit and apply scaling
        scaler = StandardScaler()
        X[numeric_cols] = scaler.fit_transform(X[numeric_cols])
    else:
        # Use pre-computed statistics from training
        if numeric_cols:
            X[numeric_cols] = scaler.transform(X[numeric_cols])
    
    if is_training:
        # Extract target variable
        y = df_processed[target_col]
        return X, y, encoder, scaler
    else:
        return X, encoder, scaler


# Step 2: Apply preprocessing to training data
print("Preprocessing training data...")
X_train, y_train, encoder, scaler = preprocess_data(train_df, is_training=True)
print(f"Training data shape: {X_train.shape}")
print(f"Target distribution:\n{y_train.value_counts()}")
print(f"\nFeature columns: {X_train.columns.tolist()}")
```

```python
# Step 3: Apply preprocessing to test data with same transformations
print("\nPreprocessing test data...")
X_test, _, _ = preprocess_data(test_df, is_training=False, encoder=encoder, scaler=scaler)
print(f"Test data shape: {X_test.shape}")

# Ensure test data has same columns as training data
missing_cols = set(X_train.columns) - set(X_test.columns)
extra_cols = set(X_test.columns) - set(X_train.columns)

if missing_cols:
    print(f"\nAdding missing columns from training: {missing_cols}")
    for col in missing_cols:
        X_test[col] = 0

if extra_cols:
    print(f"Removing extra columns not in training: {extra_cols}")
    X_test = X_test.drop(columns=extra_cols)

# Reorder test columns to match training
X_test = X_test[X_train.columns]
print(f"Final test data shape: {X_test.shape}")
```

```python
# Step 4: Retrain model on properly preprocessed data
print("\nRetraining model with preprocessed data...")
hist_boosted_v2 = sk.ensemble.HistGradientBoostingClassifier(random_state=42)
hist_boosted_v2.fit(X_train, y_train)

train_score = hist_boosted_v2.score(X_train, y_train)
print(f"Training accuracy: {train_score:.4f}")

# Make predictions on test data
predictions = hist_boosted_v2.predict(X_test)
predictions_proba = hist_boosted_v2.predict_proba(X_test)

print(f"\nPredictions shape: {predictions.shape}")
print(f"Predictions sample (first 10): {predictions[:10]}")
print(f"Prediction probabilities sample (first 5):\n{predictions_proba[:5]}")
print(f"\nChurn distribution in predictions: {np.unique(predictions, return_counts=True)}")
```

```python
# ============================================================================
# HYPERPARAMETER TUNING
# ============================================================================

from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, cross_val_score
import time

print("="*70)
print("HYPERPARAMETER TUNING FOR HISTGRADIENTBOOSTINGCLASSIFIER")
print("="*70)

# Step A: Define hyperparameter grid for HistGradientBoostingClassifier
# Note: We'll use RandomizedSearchCV for efficiency with large datasets
param_grid = {
    'learning_rate': [0.01, 0.05, 0.1, 0.15],
    'max_depth': [5, 7, 10, 15],
    'max_leaf_nodes': [20, 30, 40, 50],
    'min_samples_leaf': [10, 20, 30],
    'l2_regularization': [0.0, 0.1, 1.0],
    'max_bins': [128, 256],
}

print(f"\nTotal parameter combinations: ~{len(param_grid['learning_rate']) * len(param_grid['max_depth']) * len(param_grid['max_leaf_nodes']) * len(param_grid['min_samples_leaf']) * len(param_grid['l2_regularization']) * len(param_grid['max_bins'])}")
print("Using RandomizedSearchCV for efficient exploration...\n")

# Step B: Perform Randomized Search with cross-validation
# Sample 20 random combinations from parameter space
random_search = RandomizedSearchCV(
    estimator=sk.ensemble.HistGradientBoostingClassifier(random_state=42),
    param_distributions=param_grid,
    n_iter=20,  # Test 20 random combinations
    cv=5,  # 5-fold cross-validation
    scoring='roc_auc',  # Use AUC as metric
    n_jobs=-1,  # Use all processors
    verbose=1,
    random_state=42
)

# Fit the random search
start_time = time.time()
print("Starting hyperparameter search (this may take a few minutes)...\n")
random_search.fit(X_train, y_train)
elapsed_time = time.time() - start_time

# Step C: Display results
print(f"\nSearch completed in {elapsed_time:.2f} seconds\n")
print(f"Best CV ROC-AUC Score: {random_search.best_score_:.6f}")
print("\nBest Hyperparameters:")
for param, value in random_search.best_params_.items():
    print(f"  {param}: {value}")

# Step D: Evaluate best model vs current model
best_model = random_search.best_estimator_
best_train_score = best_model.score(X_train, y_train)

print(f"\n" + "="*70)
print("MODEL COMPARISON")
print("="*70)
print(f"Current Model Accuracy:     {train_score:.6f}")
print(f"Best Tuned Model Accuracy:  {best_train_score:.6f}")
print(f"Improvement:                {(best_train_score - train_score):.6f}")

# Step E: Show top 5 parameter combinations
print(f"\nTop 5 Parameter Combinations:")
results_df = pd.DataFrame(random_search.cv_results_)
top_results = results_df.nlargest(5, 'mean_test_score')[['param_learning_rate', 'param_max_depth', 
                                                           'param_max_leaf_nodes', 'param_min_samples_leaf',
                                                           'param_l2_regularization', 'mean_test_score']]
print(top_results.to_string())
```

```python
# Step F: Optional - Use best model for final predictions (if improvement is significant)
use_best_model = best_train_score > train_score

if use_best_model:
    print("\n" + "="*70)
    print("USING OPTIMIZED MODEL FOR FINAL PREDICTIONS")
    print("="*70)
    final_model = best_model
    final_model_name = "Tuned HistGradientBoosting"
else:
    print("\n" + "="*70)
    print("KEEPING ORIGINAL MODEL (No significant improvement found)")
    print("="*70)
    final_model = hist_boosted_v2
    final_model_name = "Original HistGradientBoosting"

print(f"\nFinal model: {final_model_name}")
print(f"Final training accuracy: {final_model.score(X_train, y_train):.6f}")

# Generate predictions with final model
predictions = final_model.predict(X_test)
predictions_proba = final_model.predict_proba(X_test)

print(f"\nPredictions generated: {predictions.shape[0]}")
print(f"Probability range: [{predictions_proba[:, 1].min():.6f}, {predictions_proba[:, 1].max():.6f}]")
```

```python
# Step 5: Create submission file with probability scores
# Extract probability of churn class (Yes) - the second column
churn_probability = predictions_proba[:, 1]

submission_df = pd.DataFrame({
    'id': test_df['id'],
    'Churn': churn_probability
})

# Save submission
submission_df.to_csv('submission.csv', index=False)
print("Submission file saved: submission.csv")
print(f"Probability range: [{churn_probability.min():.6f}, {churn_probability.max():.6f}]")
print(f"Probability mean: {churn_probability.mean():.6f}")
print(f"\nSubmission preview:\n{submission_df.head(10)}")

# ============================================================================
# PREPROCESSING PIPELINE SUMMARY
# ============================================================================
print("\n" + "="*70)
print("PREPROCESSING PIPELINE SUMMARY")
print("="*70)
print("""
The preprocessing pipeline includes the following steps:

1. DATA SEPARATION
   - Identifies numeric and categorical features
   - Separates target variable (Churn) from features
   - Excludes ID column from processing

2. CATEGORICAL ENCODING
   - One-hot encodes categorical variables
   - Uses drop_first=True to avoid multicollinearity
   - Ensures train/test columns match by adding missing columns

3. MISSING VALUE HANDLING
   - Fills numeric missing values with column mean
   - Properly handles categorical variables

4. FEATURE SCALING
   - Applies StandardScaler to numeric features
   - Ensures features are on similar scale (important for tree models)
   - Scaler is fitted on training data and applied to test data

5. SCHEMA ALIGNMENT
   - Reorders test columns to match training data
   - Adds missing columns with 0 values
   - Removes unexpected columns
   - Ensures feature consistency between train/test

INPUT: 
   - Training shape: {}, Target shape: {}
OUTPUT:
   - X_train shape: {}
   - X_test shape: {}
   - Number of features: {}

This pipeline ensures consistent preprocessing across train/test sets,
which is critical for accurate model performance on unseen data.
""".format(train_df.shape, y_train.shape, X_train.shape, X_test.shape, X_train.shape[1]))
```
